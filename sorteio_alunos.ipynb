# Sorteio de Alunos - Correção
Este notebook lê um arquivo CSV com os nomes dos alunos e sorteia 10 deles aleatoriamente, corrigindo problemas de espaços em branco nos cabeçalhos.

```python
# Importando as bibliotecas necessárias
import csv
import random

# Função para ler o arquivo CSV e remover espaços extras
def ler_alunos_csv(nome_arquivo):
    alunos = []
    with open(nome_arquivo, mode='r') as arquivo_csv:
        leitor_csv = csv.DictReader(arquivo_csv)
        leitor_csv.fieldnames = [header.strip() for header in leitor_csv.fieldnames]  # Remove espaços dos cabeçalhos
        for linha in leitor_csv:
            alunos.append(linha['alunos'].strip())  # Remove espaços extras dos nomes dos alunos
    return alunos

# Função para sortear 10 alunos
def sortear_alunos(alunos, quantidade=10):
    return random.sample(alunos, quantidade)
```

```python
# Nome do arquivo CSV
nome_arquivo = 'alunos.csv'

# Lendo os alunos
lista_alunos = ler_alunos_csv(nome_arquivo)

# Sorteando 10 alunos
alunos_sorteados = sortear_alunos(lista_alunos, 10)

# Exibindo os alunos sorteados
print("Alunos sorteados:")
for aluno in alunos_sorteados:
    print(aluno)
```
